```python
from unsloth import FastLanguageModel
import torch

model_name = "unsloth/llama-3.2-1b-bnb-4bit"

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name,
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True,
)

FastLanguageModel.for_inference(model)
```

```python
documents = [
    "TCP three-way handshake establishes a reliable connection using SYN, SYN-ACK, and ACK.",
    "Flow control in TCP ensures the sender does not overwhelm the receiver.",
    "Congestion control adjusts the data transmission rate to avoid network congestion.",
    "UDP is a connectionless protocol and does not guarantee delivery.",
]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50
)

docs = splitter.create_documents(documents)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
query = "Explain TCP three-way handshake"

retrieved_docs = vectorstore.similarity_search(query, k=2)

for doc in retrieved_docs:
    print(doc.page_content)
```

```python
context = "\n".join([doc.page_content for doc in retrieved_docs])

prompt = f"""
You are a helpful assistant.
Answer the question using ONLY the context below.

Context:
{context}

Question:
{query}

Answer:
"""
```

```python
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(
    **inputs,
    max_new_tokens=200,
    temperature=0.7,
    do_sample=True
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
